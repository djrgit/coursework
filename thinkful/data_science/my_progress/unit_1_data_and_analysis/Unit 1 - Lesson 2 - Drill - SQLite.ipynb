# Drill - SQLite

```python
import sqlite3
```

```python
file = 'bikeshare.db'
conn = sqlite3.connect(file)
c = conn.cursor()
```

### QUERY 1:  The ID's and durations for all trips of duration greater than 500, ordered by duration.

```python
q1 = c.execute('SELECT trip_id, duration \
                FROM trips \
                WHERE duration > 500 \
                ORDER BY duration')
```

```python
q1_rows = c.fetchall()
q1_rows
```

### QUERY 2:  Every column of the stations table for station id 84.

```python
q2 = c.execute('SELECT * \
                FROM stations \
                WHERE station_id = 84')
```

```python
q2_rows = c.fetchall()
q2_rows
```

### QUERY 3:  The min temperatures of all the occurrences of rain in zip 94301.

```python
q3 = c.execute('SELECT MinTemperatureF, Events \
                FROM weather \
                WHERE Events = "Rain" AND ZIP = 94301')
```

```python
q3_rows = c.fetchall()
q3_rows
```

```python
conn.close()
```
