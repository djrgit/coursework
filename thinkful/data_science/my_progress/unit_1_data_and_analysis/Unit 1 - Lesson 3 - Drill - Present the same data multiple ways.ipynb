### 1. Choose one variable and plot that variable four different ways.

###### Histogram

```python
# Showing the  information in a seaborn facet grid.
sns.set(style='whitegrid')

# Tell seaborn about the structure of our data.
g = sns.FacetGrid(wine, col='color', hue='color', palette=['red','yellow'], height=6)

# Trying (and failing) to center histogram bars on ticks
b_width = .5
my_bins = np.arange(2, 10 + b_width, b_width)

# Name the plot type and the variable to be plotted using the structure.
g.map(plt.hist, 'quality', bins=my_bins)

# Moving the plots apart to make room for the titles.
plt.subplots_adjust(wspace=.40)

# Making a more informative axis name.
g.set_axis_labels('Quality')
plt.suptitle('Plot 1: Quality Ratings for Red/White Wines')

# Removing excess lines around the plot.
sns.despine(trim=True)

plt.show()
```

###### Boxplot

```python
sns.set(style='darkgrid')
g = sns.boxplot(x='color', y='quality', palette=['red','yellow'], data=wine)
plt.title('Plot 2: Quality Ratings for Red/White Wines')
sns.despine(offset=10, trim=True)
g.set(xlabel='', ylabel='Quality')
plt.show()
```

###### Barplot

```python
g = sns.catplot(x='color', y='quality', data=wine, height=6, 
                kind='bar', palette=['red','yellow'], ci=95)
g.despine(left=True)
g.set_xlabels('')
g.set_ylabels('Quality')
plt.title('Plot 3: Average Quality Ratings for Red/White Wines')
plt.show()
```

###### Pointplot

```python
g = sns.catplot(x='color', y='quality', hue='color', palette=['red','yellow'],
                data=wine, height=6, kind='point', ci=95, join=False)
g.despine(left=True)
g.set_xlabels('')
g.set_ylabels('Quality')
plt.title('Plot 4: Average Quality Ratings for Red/White Wines')
plt.show()
```

### 2. Choose two continuous variables, and plot them three different ways.

###### Histograms

```python
g = sns.FacetGrid(wine, col='color', hue='color', palette=['red','yellow'], height=6)
g.map(plt.hist, 'fixed acidity', bins=20)
plt.subplots_adjust(wspace=.70)
g.set_axis_labels('Fixed Acidity')
plt.suptitle('Group 1, Plot 1: Fixed Acidity for Red/White Wines')
sns.despine(trim=True)

plt.show()
```

```python
g = sns.FacetGrid(wine, col='color', hue='color', palette=['red','yellow'], height=6)
g.map(plt.hist, 'residual sugar', bins=20)
plt.subplots_adjust(wspace=.70)
g.set_axis_labels('Residual Sugar')
plt.suptitle('Group 1, Plot 2: Residual Sugar for Red/White Wines')
sns.despine(trim=True)

plt.show()
```

###### Scatterplots

```python
g = sns.lmplot(y='fixed acidity', x='residual sugar', 
               data=wine, hue='color', palette=['red','yellow'],
               fit_reg=False, scatter_kws={'alpha':0.3})
plt.title('Group 2, Plot 1: Fixed Acidity/Residual Sugar Scatterplot for Red/White Wines')
plt.show()
```

```python
g = sns.lmplot(y='fixed acidity', x='residual sugar',
               data=wine, col='color', fit_reg=False)
plt.suptitle('Group 2, Plot 2')
plt.subplots_adjust(wspace=.70)
plt.show()
```

###### Boxplots

```python
g = sns.boxplot(x='color', y='fixed acidity', palette=['red','yellow'], data=wine)
plt.title('Group 3, Plot 1: Fixed Acidity for Red/White Wines')
sns.despine(offset=10, trim=True)
g.set(xlabel='', ylabel='Fixed Acidity')
plt.show()
```

```python
g = sns.boxplot(x='color', y='residual sugar', palette=['red','yellow'], data=wine)
plt.title('Group 3, Plot 2: Residual Sugar for Red/White Wines')
sns.despine(offset=10, trim=True)
g.set(xlabel='', ylabel='Residual Sugar')
plt.show()
```

###### Hexbin plot

```python
g = sns.jointplot(y='fixed acidity', x='residual sugar', data=wine, kind='hex')
plt.suptitle('Group 4, Plot 1: Residual Sugar for Red/White Wines')
plt.show()
```

### 3. Choose one continuous variable and one categorical variable, and plot them six different ways.

###### Histograms

```python
g = sns.FacetGrid(wine[['alcohol', 'color']], col='color', 
                  palette=['red','yellow'], hue='color', height=6)
g.map(plt.hist, 'alcohol', bins=20)
plt.subplots_adjust(wspace=.70)
```

###### Distribution/Rug plots

```python
g = sns.FacetGrid(wine, col="color", height=6)
g.map(sns.distplot, "alcohol", hist=False, rug=True)
plt.subplots_adjust(wspace=.70)
```

###### Boxplot

```python
g = sns.boxplot(x='color', y='alcohol', palette=['red','yellow'], data=wine)
plt.title('Plot 3: Alcohol Content in Red/White Wines')
sns.despine(offset=10, trim=True)
g.set(xlabel='', ylabel='Alcohol')
plt.show()
```

###### Barplot

```python
g = sns.catplot(x='color', y='alcohol', data=wine, height=6, 
                kind='bar', palette=['red','yellow'], ci=95)
g.despine(left=True)
g.set_xlabels('')
g.set_ylabels('Alcohol')
plt.title('Plot 4: Average Alcohol Content in Red/White Wines')
plt.show()
```

###### Pointplot

```python
g = sns.catplot(x='color', y='alcohol', hue='color', palette=['red','yellow'],
                data=wine, height=6, kind='point', ci=95, join=False)
g.despine(left=True)
g.set_xlabels('')
g.set_ylabels('Alcohol')
plt.title('Plot 5: Average Alcohol Content in Red/White Wines')
plt.show()
```

###### Swarmplots

```python
g = sns.swarmplot(x="color", y="alcohol", data=wine, palette=['red','yellow'])
plt.title('Swarm Plot of Alcohol Content in Red/White Wines')
plt.show()
```

```python
g = sns.swarmplot(x="quality", y="alcohol", data=wine)
plt.title('Swarm Plot of Alcohol Content and Quality Rating for Red/White Wines')
plt.show()
```
